```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
import pandas as pd

df = pd.read_excel("nlsw88.xlsx")
df["Hourly_wage"] = df["wage"] / df["hours"]
```

```python
df.head()
dff=df.dropna()
```

```python
dff.head()
```

```python
model = ols("Hourly_wage ~ C(race)*C(union)", data=dff).fit()
```

```python
anova_table = sm.stats.anova_lm(model, type=2)
anova_table
```
